# 3. Transfer images through a single layer

The stylized images are save in *./single_layer_compare/*

## (1) Transfer

```python
pretrained_vgg = "./vgg19.npy"
content_path = "./content/im4.jpg"
style_path = "./style/s5.jpg"
output_path = "./single_layer_compare/relu_1_1"
alpha = 0.6

tf.reset_default_graph()
transferrer = SingleLayerTransferrer(
                 target_layer = "relu1",
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 decoder_weights = 'models/decoder_1.ckpt'
                 )
transferrer.transfer()

tf.reset_default_graph()
output_path = "./single_layer_compare/relu_2_1"
transferrer = SingleLayerTransferrer(
                 target_layer = "relu2",
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 decoder_weights = 'models/decoder_2.ckpt'
                 )
transferrer.transfer()

tf.reset_default_graph()
output_path = "./single_layer_compare/relu_3_1"
transferrer = SingleLayerTransferrer(
                 target_layer = "relu3",
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 decoder_weights = 'models/decoder_3.ckpt'
                 )
transferrer.transfer()

tf.reset_default_graph()
output_path = "./single_layer_compare/relu_4_1"
transferrer = SingleLayerTransferrer(
                 target_layer = "relu4",
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 decoder_weights = 'models/decoder_4.ckpt'
                 )
transferrer.transfer()

tf.reset_default_graph()
output_path = "./single_layer_compare/relu_5_1"
transferrer = SingleLayerTransferrer(
                 target_layer = "relu5",
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 decoder_weights = 'models/decoder_5.ckpt'
                 )
transferrer.transfer()
```

## 3. Transfer images through all layers

Since we failed to train the Relu5 decoder, here we only present 4 layersstylization.

### (1) Coarse-to-fine Stylization 

```python
import time
pretrained_vgg = "./vgg19.npy"
content_path = "./content/im4.jpg"
style_path = "./style/s5.jpg"
output_path = "./multilayer_compare/"
alpha = 0.6

start = time.time()
tf.reset_default_graph()
transferrer = StyleTransferrer(
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 )
transferrer.transfer(num_layers = 4)
end = time.time()
print("Running Time:", end - start)

images = []
plt.figure()
f, axarr = plt.subplots(1, 6, figsize=(16,16))
img = utils.load_image("./style/s5.jpg", [224,224])
axarr[0].imshow(img/255)
axarr[0].set_title("Style")
img = utils.load_image("./content/im4.jpg", [224,224])
axarr[1].imshow(img/255)
axarr[1].set_title("Content")

for i in range(2,6):
    img = utils.load_image("./multilayer_compare/relu_{}_1.jpg".format(6 - i), [224,224])
    axarr[i].imshow(img/255)
    axarr[i].set_title("I{}".format(6 - i))
```

### (2) Fine-to-coarse Stylization 

```python
pretrained_vgg = "./vgg19.npy"
content_path = "./content/im4.jpg"
style_path = "./style/s5.jpg"
output_path = "./multilayer_reverse_compare/"
alpha = 0.6

tf.reset_default_graph()
transferrer = StyleTransferrer(
                 pretrained_vgg = pretrained_vgg,
                 content_path = content_path,
                 style_path = style_path,
                 output_path = output_path,
                 alpha = alpha,
                 )
transferrer.transfer(num_layers = 4, reverse = True)


images = []
plt.figure()
f, axarr = plt.subplots(1, 6, figsize=(16,16))
img = utils.load_image("./style/s5.jpg", [224,224])
axarr[0].imshow(img/255)
axarr[0].set_title("Style")
img = utils.load_image("./content/im4.jpg", [224,224])
axarr[1].imshow(img/255)
axarr[1].set_title("Content")

for i in range(2,6):
    img = utils.load_image("./multilayer_reverse_compare/relu_{}_1.jpg".format(i-1), [224,224])
    axarr[i].imshow(img/255)
    axarr[i].set_title("I{}".format(i - 1))
```
